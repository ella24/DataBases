```python
sampleBBC=[]

for row in df.iterrows():
    index, data = row
    sampleBBC.append(data.tolist())
print(sampleBBC)
```

```python
import numpy as np
import pandas as pd
col_names = ['critic_cn','critic_name','critic_org','movie_dir','movie_name','movie_rank','movie_yr','genre','dir_country']
df = pd.DataFrame.from_records(sampleBBC, columns=col_names)
df.loc[20]
```

```python
df.groupby('critic_cn')['critic_name'].nunique()
```

```python
df1 = df.groupby('critic_cn')['movie_name'].value_counts().reset_index(name='count')
df1.head(40)
```

```python
df1 = df.groupby('critic_cn')['dir_country'].value_counts().reset_index(name='count')
df1.head(40)
```

```python
df1['string'] = df1['movie_name'] + ': ' + df1['count'].map(str) + np.where(df1['count']>1, ' votes', ' vote')
df1.head()
```

```python
output = df1.groupby('critic_cn')['string'].apply(lambda x: "<p>%s</p>" % '</p><p> '.join(x)).reset_index(name='properties.article')
output
```

```python
output.iloc[3]['properties.article']
```

```python
critics = df.groupby('critic_cn')['critic_name'].nunique().reset_index(name='properties.headline')
output = output.merge(critics, how='left', on='critic_cn')
output
```

```python
output['properties.headline'] = output['properties.headline'].map(str) + ' critic_name'
output
```
